# Değişken Standardizasyonu (Veri Standardizasyonu)

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,5,7])
V2 = np.array([7,7,5,8,12])
V3 = np.array([6,12,5,6,14])
df = pd.DataFrame(
        {"V1" : V1,
         "V2" : V2,
         "V3" : V3})

df = df.astype(float)
df
```

## Standardizasyon

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df) #dönüştürme işlemi yapmış olduk
```

```python
df
```

## Normalizasyon

```python
preprocessing.normalize(df) # değerleri 0 ile 1 arasında değişexek şekilde dönüştürme işlemi yapar
```

## Min - Max Dönüşüm

```python
scaler = preprocessing.MinMaxScaler(feature_range = (10,20))  # kendi belirlediğimiz aralıklarda dönüştürme işlemi yapmak istersek
```

```python
scaler.fit_transform(df)
```

# Değişken Dönüştürme

```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

#### 0-1 Dönüşümü 

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
```

```python
yeni_sex = lbe.fit_transform(df["sex"])
```

```python
df["yeni_sex"] = yeni_sex
```

```python
df
```

#### 1 ve Diğerleri(0) Dönüşümü

```python
df.head()
```

```python
import numpy as np
df["yeni_day"] = np.where(df["day"].str.contains("Sun"),1,0)# day içerisinde sun değerleri içeren gözlem birimlerine 1 diğerlerine 0 değerlerini atar
```

```python
df
```
