# Çok Değişkenli Aykırı Gözlem Analizi

### Local Outlier Factor

Gözlemleri bulundukları konumda yoğunluk tabanlı skorlayarak buna göre aykırı değe olabilecek değerler tanımlayabilmemize imkan sağlıyor

Bir noktanın local yoğunluğu bu noktanın komşuları ile karşılaştırılıyor. Eğer bir nokta komşularının yoğunluğunda anlamlı şekilde düşük ise bu nokta komşularından daha seyrek bir bölgede bulunuyordur yorumu yapılabiliyor.Dolayısıyla burada bir komşuluk yapısı söz konusu.Bir değerin çevresi yoğun değilse demek ki bu değer aykırı değerdir şeklinde değerlendiriliyor.

```python
import seaborn as sns
df = sns.load_dataset('diamonds')
df = df.select_dtypes(include = ['float64','int64'])
df = df.dropna()
df.head(10)
```

```python
import numpy as np
```

```python
from sklearn.neighbors import LocalOutlierFactor # sklearn kütüphanesinin içerisinden neighbors içerisinden LocalOutlierFactor fonksiyonunu import ettik(sklearn kütüphanesinin neighbors modülünde Local Outlier Factor sınıfı bulunmaktadır.) 
```

```python
clf = LocalOutlierFactor(n_neighbors = 20, contamination = 0.1) #komşuluk ve yoğunluk argümanları
```

```python
clf.fit_predict(df)  # 1 normal değerler -1 ise aykırı değerleri veriyor
```

```python
df_scores = clf.negative_outlier_factor_
```

```python
df_scores[0:20]   #her bir değerin yoğunluk skorları 
```

```python
np.sort(df_scores)[0:20]   
```

```python
np.sort(df_scores)[13]
```

##### silme yöntemi

```python
esik_deger = np.sort(df_scores)[13] #skorların içerisindeki 13. indexteki değeri eşik değeri olarak belirledik
```

```python
aykiri_tf = df_scores > esik_deger # galiba aykırı olmayan değerler
```

```python
aykiri_tf
```

```python
yeni_df = df[aykiri_tf]   # aykırı olmayan değerlere eriştik
```

```python
yeni_df
```

#### Baskılama Yöntemi

```python
df[df_scores == esik_deger]  # belirlediğimiz eşik değerin numerik gözlem birimindeki karşılığını gösteriyor
```

```python
baski_deger = df[df_scores == esik_deger]
```

```python
aykirilar = df[~aykiri_tf]
```

```python
aykirilar
```

```python
aykirilar.to_records(index = False)# bunu numpy arrrayıne çevirdi indexlerinden kurtulmuş olduk çünkü indexlerin sıkıntı çıkarmasını istemiyoruz
```

```python
res = aykirilar.to_records(index = False)# baştan bir index oluşturma işlemi
```

```python
res[:] = baski_deger.to_records(index = False)  # aykiri değerlerimizin hepsini baski değerlerine eşitlemiş olduk, baski değer dediğim şey ise belirlemiş olduğumuz eşik değer
res
```

```python
df[~aykiri_tf]  # aykirilarda bir değişiklik olmamış bu yüzden yukardaki arrayi bu veri setimize yerleştirmemiz gerekiyor
```

```python
import pandas as pd
df[~aykiri_tf] = pd.DataFrame(res,index = df[~aykiri_tf].index) # aykiri_tf nin indexlerini kullandık
```
